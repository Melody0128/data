```python
#coding:utf-8
import os
import pandas as pd
import numpy as np
```

# read

```python
policy_all = pd.DataFrame()
dire = 'D:\\a latest paper\\first\\policy\\data\\data\\policy20112021\\'
for root, dirs, files in os.walk(dire):
    for file_single in files:
        path = root+'/'+file_single
        path = path.replace('\\', '/')
        # refile = file_single #[0:8]
        # result = path +'\n'
        # print(path)
        with open(path, 'r', encoding='UTF-8') as f:  # 设置文件对象
            print(path)
            text = f.read()
            temp = pd.DataFrame(columns=['path', 'content'])
            temp.loc[0, 'path'] = path
            temp.loc[0, 'content'] = text
            # print(temp)
            # print(text)
            policy_all = policy_all.append(temp, ignore_index=True)
            policy_all.to_csv('policy_all.csv', encoding='utf_8_sig', index=False)
```

```python
policy_all.to_csv('policy_all.csv', encoding='utf_8_sig', index=False)
```

# Accession 'year', 'province'

```python
for i in range(len(policy_all['path'])):
    policy_all.loc[i, 'province'] = policy_all['path'][i].split('/')[5]

    policy_all.loc[i, 'year'] = '20'+policy_all['path'][i].split('/')[6]

    temp_title = policy_all['path'][i].split('/')[7][:-4]
    if temp_title[:4].isdigit():
        temp_title = temp_title[4:]
    elif temp_title[:2].isdigit():
        temp_title = temp_title[2:]
    policy_all.loc[i, 'title'] = temp_title

    policy_all.loc[i, 'ID'] = str(i)
```

## Categorized Tags

```python
def ConvertCats(d):
    if d< 5:
        return '无关'
    elif 5<= d < 20:
        return '提及'
    elif 20<= d < 50:
        return '涉及'
    else:
        return '强调'
```

```python
policy_cats_cats= pd.DataFrame(np.zeros((len(policy_cats_value),20)),
                               columns =['A_cats','B_cats','C_cats','D_cats','E_cats',
                                         'F_cats','G_cats','H_cats','I_cats','J_cats',
                                         'K_cats','L_cats','M_cats','N_cats','O_cats',
                                         'P_cats','Q_cats','R_cats','S_cats','T_cats'])
policy_cats_cats
```

```python
for i in range(policy_cats_cats.shape[0]):
    for j in range(policy_cats_cats.shape[1]):
        policy_cats_cats.iloc[i, j] = ConvertCats(policy_cats_value.iloc[i, j])
policy_cats_cats        
```
